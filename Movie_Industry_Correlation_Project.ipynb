Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12,8) #adjusts configuration of the plots that will be created
```

Read the Data File into Python using Pandas

```python
path = '/content/movies.csv'
movie_ind = pd.read_csv(path)
#look at the data file
movie_ind.head()
```

Look for Missing Data

```python
for col in movie_ind.columns:
  percent_missing = np.mean(movie_ind[col].isnull())
  print("{} - {}%".format(col,percent_missing))
```

```python
#sums up all the blank spaces found within each column.
movie_ind.isna().sum()
```

Data Cleaning

```python
#mathematical information based on values found in the data file.
movie_ind.describe()
```

```python
#looking at the types of values found within each column.
movie_ind.dtypes
```

```python
movie_ind = movie_ind.sort_values(by=["gross"],inplace = False, ascending=False)
```

```python
#method to see all the values in the data table
pd.set_option("display.max_rows",None)
```

```python
movie_ind["company"].drop_duplicates().sort_values(ascending=False)
```

Finding Correlations in the Data

```python
#scatter plot budget vs gross revenue
plt.scatter(x=movie_ind["budget"], y=movie_ind["gross"])
plt.title("Budget Vs Gross Revenue")
plt.xlabel("Budget for Film(in million $)")
plt.ylabel("Gross Revenue(in million $)")
plt.show()
```

```python
movie_ind.head()
```

```python
#plot budget vs gross using seaborn package
sns.regplot(x="budget",y="gross", data = movie_ind, scatter_kws={"color":"orange"},line_kws={"color":"black"});
```

```python
#pearson correlation
movie_ind.corr(method="pearson")
```

```python
#heatmap of the correlation matrix
from scipy.spatial.distance import correlation
correlation_matrix = movie_ind.corr(method="pearson")
sns.heatmap(correlation_matrix, annot = True, cmap="coolwarm");
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show();
```

```python
#looks at company
movie_ind.head()
```

```python
#Assign each company with a number for calculations
MovInd_Numerized = movie_ind
for col_name in MovInd_Numerized.columns:
  if(MovInd_Numerized[col_name].dtype == "object"):
    MovInd_Numerized[col_name] = MovInd_Numerized[col_name].astype("category")
    MovInd_Numerized[col_name] = MovInd_Numerized[col_name].cat.codes

MovInd_Numerized.head()
```

```python
movie_ind.head()
```

```python
from scipy.spatial.distance import correlation
correlation_matrix = MovInd_Numerized.corr(method="pearson")
sns.heatmap(correlation_matrix, annot = True, cmap="coolwarm");
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show();
```

```python
MovInd_Numerized.corr()
```

```python
corr_mat = MovInd_Numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_corr_pairs = corr_pairs.sort_values()
sorted_corr_pairs
```

```python
high_corr = sorted_corr_pairs[(sorted_corr_pairs)>= 0.5]
high_corr
#budget and votes had the highest correlation on the gross revenue earned
```
